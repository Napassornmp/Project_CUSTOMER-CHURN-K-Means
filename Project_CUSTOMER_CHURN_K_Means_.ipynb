```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
```

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

categorical_cols = df_features.columns

for col in categorical_cols:
    print(f"{col}: {cramers_v(df_features['cluster'], df_features[col]):.3f}")


```

## Cramér’s V หากมีค่าสูงๆ แปลว่า Feature นั้นมีอิทธิพลต่อการจัดกลุ่มมาก
